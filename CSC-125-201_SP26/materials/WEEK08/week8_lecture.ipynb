### Basic For Loops

A for loop repeats a block of code once for each item in a sequence. The syntax uses a colon and indentation, just like `if` statements.

```python
for item in sequence:
    # code to run for each item
```

````python
# Simple for loop through a list
fruits = ["apple", "banana", "cherry", "date"]

print("Fruits in my list:")
for fruit in fruits:
    print(fruit)

print("\nDone printing fruits.")
````

````python
# The loop variable name is your choice
# Use a name that describes what each item represents
prices = [4.99, 2.50, 8.75, 1.25]

for price in prices:
    print(f"Item costs ${price:.2f}")
````

````python
# Combining loops with conditionals
scores = [85, 92, 78, 95, 88, 61, 73]

print("Score Report:")
for score in scores:
    if score >= 90:
        label = "Excellent"
    elif score >= 80:
        label = "Good"
    elif score >= 70:
        label = "Satisfactory"
    else:
        label = "Needs improvement"
    print(f"  {score}: {label}")
````

### Building New Lists with Loops

Start with an empty list and use `.append()` inside the loop to build up a result. This is one of the most common loop patterns.

````python
# Calculate sales tax for each price
prices = [10.99, 25.50, 15.75, 8.25]
tax_rate = 0.08

prices_with_tax = []
for price in prices:
    taxed = price * (1 + tax_rate)
    prices_with_tax.append(taxed)

print(f"Original: {prices}")
print(f"With tax: {prices_with_tax}")
````

````python
# Filter items: keep only values that match a condition
numbers = [3, 8, 15, 6, 22, 9, 18, 5]

large_numbers = []
for num in numbers:
    if num > 10:
        large_numbers.append(num)

print(f"All numbers: {numbers}")
print(f"Numbers > 10: {large_numbers}")
````

### The range() Function

Sometimes you need to repeat something a specific number of times, or count through a sequence of numbers. The `range()` function generates numbers on demand.

````python
# range(stop) - counts from 0 up to (but not including) stop
print("Counting 0 to 4:")
for i in range(5):
    print(i)

print()

# Useful for repeating something N times
print("Repeating 3 times:")
for i in range(3):
    print("Hello.")
````

````python
# range(start, stop) - counts from start up to stop - 1
print("Counting 1 to 5:")
for i in range(1, 6):
    print(i)

print()

# Practical example: year range
print("Recent years:")
for year in range(2022, 2027):
    print(year)
````

````python
# range(start, stop, step) - counts by a custom amount
print("Even numbers 0 to 10:")
for i in range(0, 11, 2):
    print(i)

print()

# Counting backward with a negative step
print("Countdown:")
for i in range(10, 0, -1):
    print(i)
print("Go.")
````

### Using range() with List Indices

When you need to know the position of each item, or you need to modify a list in place, use `range(len(list))` to loop through indices.

````python
# Numbering items with their position
students = ["Alice", "Bob", "Charlie", "Diana"]

print("Class Roster:")
for i in range(len(students)):
    print(f"{i + 1}. {students[i]}")
````

````python
# Modifying list items in place
scores = [85, 92, 78, 95, 88]
print(f"Original scores: {scores}")

# Add 5 bonus points to every score
for i in range(len(scores)):
    scores[i] = scores[i] + 5

print(f"With bonus:      {scores}")
````

````python
# Processing parallel lists (same index = related data)
names = ["Alice", "Bob", "Charlie", "Diana"]
scores = [85, 92, 78, 95]

print("Student Scores:")
for i in range(len(names)):
    print(f"  {names[i]}: {scores[i]}")
````

### Looping Through Strings

Strings are sequences of characters, so a for loop can walk through each character one at a time. This is useful for counting specific characters, filtering, or building new strings.

````python
# Loop through each character
word = "Python"

print("Letters in 'Python':")
for letter in word:
    print(letter)
````

````python
# Count specific characters
text = "Hello World"
vowel_count = 0

for char in text:
    if char.lower() in "aeiou":
        vowel_count = vowel_count + 1

print(f"Text: {text}")
print(f"Number of vowels: {vowel_count}")
````

````python
# Build a new string character by character
text = "Hello 123 World"
letters_only = ""

for char in text:
    if char.isalpha():
        letters_only = letters_only + char

print(f"Original: {text}")
print(f"Letters only: {letters_only}")
````

### Common Patterns

For loops follow a few recurring patterns. Learning to recognize them makes writing new loops much easier.

#### Pattern 1: Accumulating a Total

````python
# Sum up all prices (the cash register pattern)
prices = [12.99, 8.50, 15.75, 22.00, 6.25]

total = 0
for price in prices:
    total = total + price

print(f"Prices: {prices}")
print(f"Total: ${total:.2f}")
print(f"Average: ${total / len(prices):.2f}")
````

#### Pattern 2: Counting Matches

````python
# Count passing and failing scores
scores = [85, 92, 55, 78, 95, 63, 88, 45, 91]

passing = 0
failing = 0

for score in scores:
    if score >= 60:
        passing = passing + 1
    else:
        failing = failing + 1

print(f"Total tests: {len(scores)}")
print(f"Passing: {passing}")
print(f"Failing: {failing}")
print(f"Pass rate: {passing / len(scores) * 100:.1f}%")
````

#### Pattern 3: Finding Maximum with Position

````python
# Find the highest score and who earned it
students = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
scores = [85, 92, 78, 95, 88]

# Start by assuming the first item is the best
highest_score = scores[0]
top_student = students[0]

for i in range(len(students)):
    if scores[i] > highest_score:
        highest_score = scores[i]
        top_student = students[i]

print(f"Top student: {top_student}")
print(f"Highest score: {highest_score}")
````

#### Pattern 4: Processing Parallel Lists

````python
# Shopping cart with items, prices, and quantities
items = ["apple", "banana", "orange", "grape"]
prices = [1.50, 0.75, 2.00, 3.50]
quantities = [3, 5, 2, 1]

print("Shopping Cart:")
total_cost = 0

for i in range(len(items)):
    item_total = prices[i] * quantities[i]
    total_cost = total_cost + item_total
    print(f"  {quantities[i]} {items[i]} @ ${prices[i]:.2f} = ${item_total:.2f}")

print(f"\nTotal: ${total_cost:.2f}")
````

### Nested Loops

A loop inside another loop is called a nested loop. The inner loop runs completely for every single step of the outer loop.

````python
# See how nested loops execute
for i in range(1, 4):
    print(f"Outer loop: i = {i}")
    for j in range(1, 4):
        print(f"  Inner loop: j = {j}, product = {i * j}")
    print()
````

````python
# Grid of coordinates
print("Grid:")
for row in range(3):
    for col in range(4):
        print(f"({row},{col})", end=" ")
    print()  # new line after each row
````

````python
# Generate all combinations from two lists
colors = ["red", "blue", "green"]
sizes = ["small", "medium", "large"]

print("All combinations:")
for color in colors:
    for size in sizes:
        print(f"  {size} {color} shirt")
````

````python
# Pattern: right triangle of stars
print("Right Triangle:")
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()

print()

# 5x5 Multiplication table
print("Multiplication Table:")
for i in range(1, 6):
    for j in range(1, 6):
        product = i * j
        print(f"{product:4}", end="")
    print()
````

### Common Mistakes

````python
# Mistake 1: Off-by-one with range()
# range() stops BEFORE the stop number
print("range(1, 5) gives:")
for i in range(1, 5):
    print(i, end=" ")
print("  <-- 5 is NOT included")

print()

# To include 5, use 6 as the stop
print("range(1, 6) gives:")
for i in range(1, 6):
    print(i, end=" ")
print("  <-- now 5 IS included")
````

````python
# Mistake 2: Indentation determines what is inside the loop
total = 0
numbers = [10, 20, 30]

# WRONG: print inside the loop runs every iteration
print("Wrong (prints 3 times):")
for num in numbers:
    total = total + num
    print(f"Total so far: {total}")

print()

# RIGHT: print outside the loop runs once at the end
total = 0
print("Right (prints once):")
for num in numbers:
    total = total + num
print(f"Final total: {total}")
````

````python
# Mistake 3: Modifying a list while looping through it
# This causes items to be skipped.

# WRONG approach:
numbers = [1, 2, 3, 4, 5, 6]
# for num in numbers:
#     if num % 2 == 0:
#         numbers.remove(num)  # skips items

# CORRECT approach: build a new list
numbers = [1, 2, 3, 4, 5, 6]
odd_only = []
for num in numbers:
    if num % 2 != 0:
        odd_only.append(num)

print(f"Original: {numbers}")
print(f"Odd only: {odd_only}")
````

### Complete Example: Grade Analyzer

````python
# Complete Example: Analyze a class of test scores
print("=== Grade Analyzer ===")
print()

scores = [85, 92, 78, 95, 88, 73, 91, 67, 82, 90]
print(f"Scores: {scores}")
print()

# Step 1: Calculate average
total = 0
for score in scores:
    total = total + score
average = total / len(scores)

# Step 2: Count grades by category
a_count = 0
b_count = 0
c_count = 0
d_count = 0
f_count = 0

for score in scores:
    if score >= 90:
        a_count = a_count + 1
    elif score >= 80:
        b_count = b_count + 1
    elif score >= 70:
        c_count = c_count + 1
    elif score >= 60:
        d_count = d_count + 1
    else:
        f_count = f_count + 1

# Step 3: Build a list of scores that need improvement
needs_help = []
for score in scores:
    if score < 75:
        needs_help.append(score)

# Step 4: Display results
print("Statistics:")
print(f"  Average: {average:.1f}")
print(f"  Highest: {max(scores)}")
print(f"  Lowest:  {min(scores)}")
print()

print("Grade Distribution:")
print(f"  A's: {a_count}")
print(f"  B's: {b_count}")
print(f"  C's: {c_count}")
print(f"  D's: {d_count}")
print(f"  F's: {f_count}")
print()

print(f"Scores below 75: {needs_help}")
if len(needs_help) > 0:
    print(f"  {len(needs_help)} student(s) may need extra support.")
else:
    print("  All students scored 75 or above.")
````
