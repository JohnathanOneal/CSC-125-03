### Why Lists?

Up to now, each piece of data needs its own variable. If you have 5 test scores, you need 5 variables. What about 50? 500? Lists solve this by holding multiple values in a single container.

````python
# The old way: one variable per value
score1 = 85
score2 = 92
score3 = 78
score4 = 95
score5 = 88

# Tedious to work with all of them
print(f"Scores: {score1}, {score2}, {score3}, {score4}, {score5}")
print()

# The list way: one variable, all values
scores = [85, 92, 78, 95, 88]
print(f"Scores: {scores}")
print(f"Number of scores: {len(scores)}")
print(f"Average: {sum(scores) / len(scores):.1f}")
````

---

### Creating Lists

Lists are created with square brackets `[]`. Items are separated by commas. Lists can hold any type of data, and an empty list is just `[]`.

````python
# String list
colors = ["red", "blue", "green", "yellow"]
print(f"Colors: {colors}")

# Number list
temperatures = [72, 68, 75, 80, 65]
print(f"Temperatures: {temperatures}")

# Mixed types (allowed, but usually keep items the same type)
profile = ["Alice", 25, 3.8, True]
print(f"Profile: {profile}")

# Empty list
todo = []
print(f"Empty list: {todo}")
print(f"Length: {len(todo)}")
````

### Indexing

Every item in a list has a position number called an **index**. Python starts counting at 0, not 1. Negative indices count backward from the end of the list.

````python
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter"]

print(f"List: {planets}")
print()

# Positive indexing (from the front, starting at 0)
first_planet = planets[0]
third_planet = planets[2]
print(f"Index 0: {first_planet}")
print(f"Index 2: {third_planet}")
print()

# Negative indexing (from the end, starting at -1)
last_planet = planets[-1]
second_to_last = planets[-2]
print(f"Index -1: {last_planet}")
print(f"Index -2: {second_to_last}")
````

**Visual:**
```
Position:   ["Mercury", "Venus", "Earth", "Mars", "Jupiter"]
Positive:       0          1        2        3         4
Negative:      -5         -4       -3       -2        -1
```

### Lists are Mutable

Lists can be changed after creation. You can update any item by assigning a new value to its index. This property is called **mutability**.

````python
menu = ["coffee", "tea", "juice", "water"]
print(f"Original menu: {menu}")

# Change an item
menu[2] = "smoothie"
print(f"Updated menu:  {menu}")

# The change happens in place - no new list is created
````

### List Methods

Lists come with built-in methods to add, remove, and inspect items. Here are the most commonly used ones.

````python
# Building a grocery list
grocery = []
print(f"Start: {grocery}")

# .append() adds to the end
grocery.append("bread")
grocery.append("milk")
grocery.append("eggs")
grocery.append("butter")
print(f"After appending: {grocery}")

# .remove() removes the first occurrence of a value
grocery.remove("milk")
print(f"After removing milk: {grocery}")

# .pop() removes and returns the last item
last_item = grocery.pop()
print(f"Popped: {last_item}")
print(f"Remaining: {grocery}")

# .pop(index) removes at a specific position
first_item = grocery.pop(0)
print(f"Popped index 0: {first_item}")
print(f"Remaining: {grocery}")
````

````python
prices = [4.99, 2.50, 8.75, 1.25, 6.00]
print(f"Prices: {prices}")
print()

# Built-in functions that work with lists
print(f"Count:   {len(prices)}")
print(f"Highest: ${max(prices):.2f}")
print(f"Lowest:  ${min(prices):.2f}")
print(f"Total:   ${sum(prices):.2f}")
print(f"Average: ${sum(prices) / len(prices):.2f}")
````

### Tuples

Tuples are similar to lists but with one key difference: they **cannot be changed** after creation. This property is called **immutability**. Tuples are created with parentheses `()` instead of square brackets. Use tuples when your data should remain fixed. Tuples cannot be modified after creation. Attempting to change an item raises a `TypeError`.

````python
# Creating tuples
coordinates = (42.36, -71.06)
rgb_red = (255, 0, 0)
student = ("Maria", 21, "Computer Science")

print(f"Coordinates: {coordinates}")
print(f"RGB Red: {rgb_red}")
print(f"Student: {student}")
print()

# Indexing works the same as lists
latitude = coordinates[0]
longitude = coordinates[1]
print(f"Latitude: {latitude}")
print(f"Longitude: {longitude}")
print()

# Length works too
print(f"Student has {len(student)} fields")
````

````python
# Tuple unpacking: assign each item to a variable in one line
name, age, major = student
print(f"{name} is {age} years old, studying {major}")
print()

# Unpacking coordinates
x, y = coordinates
print(f"x = {x}, y = {y}")
print()

# Swapping variables with tuple unpacking
a = 5
b = 10
print(f"Before swap: a = {a}, b = {b}")
a, b = b, a
print(f"After swap:  a = {a}, b = {b}")
````

### Lists vs Tuples

| Feature | List | Tuple |
|---------|------|-------|
| Syntax | `[1, 2, 3]` | `(1, 2, 3)` |
| Mutable | Yes | No |
| Use when | Data will change | Data should stay fixed |
| Examples | Shopping list, scores | Coordinates, colors, dates |

### Introduction to Pseudocode

Pseudocode is structured English that describes a program's logic step by step. Think of it as a blueprint before building a house, or a recipe before cooking. You write out what the program should do in plain language before translating it to Python.

**Why bother?** Writing pseudocode first helps you focus on the logic without getting distracted by syntax errors, colons, and indentation. It is much easier to fix a logic mistake in plain English than in code.

This week's homework is a pseudocode design document, so we will practice this skill in class today.

### Pseudocode Conventions

Start each step with a clear **action verb**:

| Verb | Meaning | Python equivalent |
|------|---------|-------------------|
| GET / ASK | Receive input | `input()` |
| CREATE / SET | Store a value or make a list | `x = value` or `x = []` |
| CALCULATE | Do math | `result = a + b` |
| FIND | Locate a value in data | `max(list)`, `min(list)` |
| DISPLAY | Show output | `print()` |
| IF / ELSE | Make a decision | `if / else` |
| FOR EACH | Repeat for every item | `for item in list` |

**Rules of thumb:**
- Number your steps
- Be specific: "CALCULATE the average by dividing the total by the number of scores" not just "find the average"
- Use plain language a non-programmer could follow
- Do not write Python syntax. The goal is to think through the logic, not write code

### Example: From Problem to Pseudocode to Python

**Problem:** A store has a list of daily sales amounts. Find the total sales, the best day, and whether the weekly goal of $500 was met.

**Step 1: Write the pseudocode**

````python
# PSEUDOCODE:
# 1. CREATE a list of daily sales amounts
# 2. CALCULATE the total by finding the sum of the sales list
# 3. FIND the highest value in the sales list (best day)
# 4. SET the weekly goal to 500
# 5. IF total is greater than or equal to the goal
# 6.     DISPLAY "Goal met" and the total
# 7. ELSE
# 8.     CALCULATE the shortfall as goal minus total
# 9.     DISPLAY "Goal not met" and the shortfall
# 10. DISPLAY the best single-day sales amount

# PYTHON TRANSLATION:

# 1. CREATE a list of daily sales amounts
daily_sales = [95, 120, 45, 110, 88, 72, 55]

# 2. CALCULATE the total
total_sales = sum(daily_sales)

# 3. FIND the best day
best_day = max(daily_sales)

# 4. SET the weekly goal
goal = 500

# 5-9. Check if goal was met
if total_sales >= goal:
    print(f"Goal met. Total sales: ${total_sales}")
else:
    shortfall = goal - total_sales
    print(f"Goal not met. Short by: ${shortfall}")

# 10. DISPLAY best day
print(f"Best single-day sales: ${best_day}")
````

### Complete Example: Student Grade Book

````python
# Complete Example: Student Grade Book

# Setup
student_name = "Alex"
assignments = [88, 75, 92, 84, 96]
quizzes = [90, 85]

print(f"Student: {student_name}")
print(f"Assignments: {assignments}")
print(f"Quizzes: {quizzes}")
print()

# Calculate averages
assignment_avg = sum(assignments) / len(assignments)
quiz_avg = sum(quizzes) / len(quizzes)
print(f"Assignment Average: {assignment_avg:.1f}")
print(f"Quiz Average: {quiz_avg:.1f}")
print()

# Overall grade (assignments 70%, quizzes 30%)
overall = (assignment_avg * 0.70) + (quiz_avg * 0.30)
print(f"Overall Grade: {overall:.1f}")

# Letter grade
if overall >= 90:
    letter = "A"
elif overall >= 80:
    letter = "B"
elif overall >= 70:
    letter = "C"
elif overall >= 60:
    letter = "D"
else:
    letter = "F"

print(f"Letter Grade: {letter}")
print()

# Add a new assignment
assignments.append(91)
new_avg = sum(assignments) / len(assignments)
print(f"After new assignment: {assignments}")
print(f"New Assignment Average: {new_avg:.1f}")
````
