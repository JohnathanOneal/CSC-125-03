## Part 4: Python as a Calculator

## Part 6: Working with Text (Strings)

Python can also work with text. Text in quotes is called a **string**.

Run these examples:

```python
# You can use single or double quotes
print("Hello, world!")
print('Hello, world!')
```

```python
# You can "add" strings together (concatenation)
print("Hello" + "World")
```

```python
# Add a space between words
print("Hello " + "World")
```

```python
# Using commas adds spaces automatically
print("Hello", "World")
```

```python
# You can repeat strings
print("Ha" * 5)
```
