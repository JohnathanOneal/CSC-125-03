### if-else Statements

Sometimes you want to do one thing if a condition is true and a **different** thing if it is false. That is where `if-else` comes in.

**Syntax:**
```python
if condition:
    # runs when condition is True
else:
    # runs when condition is False
```

**Key point:** Exactly one block always runs. Either the `if` block or the `else` block, never both, never neither.

````python
# Even/odd checker
number = 17

if number % 2 == 0:
    parity = "even"
    print(f"{number} is even.")
else:
    parity = "odd"
    print(f"{number} is odd.")

print(f"Result: {parity}")
````

````python
# Pass/fail checker
score = 68

if score >= 70:
    status = "PASS"
    print("You passed the course.")
else:
    status = "FAIL"
    print("You did not pass. Please speak with the instructor.")

print(f"Final status: {status}")
````

### if-elif-else Chains

When you have more than two possible outcomes, use `elif` (short for "else if") to create a chain of conditions.

**Syntax:**
```python
if condition1:
    # runs if condition1 is True
elif condition2:
    # runs if condition2 is True (and condition1 was False)
elif condition3:
    # runs if condition3 is True (and previous conditions were False)
else:
    # runs if none of the above were True
```

**Critical rule:** Python checks conditions **top to bottom** and stops at the **first one that is True**. Only one block in the chain ever runs.

````python
# Letter grade calculator
score = 87

if score >= 90:
    letter_grade = "A"
    message = "Excellent work."
elif score >= 80:
    letter_grade = "B"
    message = "Good job."
elif score >= 70:
    letter_grade = "C"
    message = "Satisfactory."
elif score >= 60:
    letter_grade = "D"
    message = "Needs improvement."
else:
    letter_grade = "F"
    message = "Please see instructor."

print(f"Score: {score}")
print(f"Grade: {letter_grade}")
print(f"Comment: {message}")
````

### Logical Operators

Sometimes a single comparison is not enough. You need to combine multiple conditions. Python provides three **logical operators**:

| Operator | Meaning | True when... |
|----------|---------|--------------|
| `and` | Both must be True | `True and True` |
| `or` | At least one must be True | `True or False`, `False or True`, `True or True` |
| `not` | Flips the value | `not False` |

````python
# Using 'and': both conditions must be True
# Driving example: must be old enough AND have a license
age = 22
has_license = True

print(f"Age: {age}, Has license: {has_license}")

if age >= 16 and has_license:
    print("You can drive.")
else:
    print("You cannot drive.")
````

````python
# Using 'or': at least one condition must be True
# Day off example: weekend OR holiday
is_weekend = False
is_holiday = True

print(f"Is weekend: {is_weekend}, Is holiday: {is_holiday}")

if is_weekend or is_holiday:
    print("You have the day off.")
else:
    print("It is a work day.")
````

````python
# Using 'not': flips True to False and False to True
is_logged_in = False

print(f"Logged in: {is_logged_in}")

if not is_logged_in:
    print("Please log in to continue.")
else:
    print("Welcome back.")
````

### Nested Conditionals

Sometimes a decision depends on a previous decision. You can place an `if` statement **inside** another `if` statement. This is called **nesting**.

Each level of nesting adds another 4 spaces of indentation.

**Tip:** Try to keep nesting to 2 levels maximum. Deeply nested code is hard to read and debug.

````python
# Movie ticket pricing
# Step 1: Determine age category
# Step 2: For adults, check if student for a discount
age = int(input("Enter your age: "))

if age < 13:
    price = 8
    print(f"Child ticket: ${price}")
elif age >= 65:
    price = 9
    print(f"Senior ticket: ${price}")
else:
    # Adult pricing: check for student discount
    is_student_input = input("Are you a student? (yes/no): ")
    is_student = is_student_input.lower() == "yes"

    if is_student:
        price = 10
        print(f"Adult student ticket: ${price}")
    else:
        price = 14
        print(f"Adult regular ticket: ${price}")

print(f"\nTotal: ${price}")
````
