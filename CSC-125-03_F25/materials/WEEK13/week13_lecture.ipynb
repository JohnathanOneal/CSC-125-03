## 1. What Are Modules?

A **module** is a file with Python code (functions, variables) that someone else wrote.

**Why use modules?**
- Do things you can't easily do yourself (random numbers, dates)
- Save time - don't reinvent the wheel
- Code that's already tested and reliable

```python
# To use a module, import it first
import random

# Then use its functions
number = random.randint(1, 10)
print(f"Random number: {number}")
```

## 2. The Random Module

**Problem:** How do you make truly random choices in Python?

You can't! You need the `random` module.

```python
import random

# Random integer between two numbers
dice = random.randint(1, 6)
print(f"Dice roll: {dice}")
print()

# Random choice from a list
options = ["Rock", "Paper", "Scissors"]
computer_choice = random.choice(options)
print(f"Computer chose: {computer_choice}")
print()

# Shuffle a list (changes it in place)
deck = ["A", "K", "Q", "J", "10"]
print(f"Before: {deck}")
random.shuffle(deck)
print(f"After shuffle: {deck}")
```

```python
# Practical example: Rock Paper Scissors
import random

def play_rps(player_choice):
    choices = ["rock", "paper", "scissors"]
    computer = random.choice(choices)
    
    print(f"You chose: {player_choice}")
    print(f"Computer chose: {computer}")
    
    if player_choice == computer:
        return "Tie!"
    elif (player_choice == "rock" and computer == "scissors") or \
         (player_choice == "paper" and computer == "rock") or \
         (player_choice == "scissors" and computer == "paper"):
        return "You win!"
    else:
        return "Computer wins!"

result = play_rps("rock")
print(result)
```

```python
# Another example: Random name picker
import random

def pick_random_student(students):
    chosen = random.choice(students)
    return chosen

class_list = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
volunteer = pick_random_student(class_list)
print(f"Today's volunteer: {volunteer}")
```

## 3. The Datetime Module

**Problem:** How do you work with dates and times?

Dates are complicated! (leap years, different month lengths, time zones...)

The `datetime` module handles all that complexity.

```python
import datetime

# Get today's date
today = datetime.date.today()
print(f"Today: {today}")
print(f"Year: {today.year}")
print(f"Month: {today.month}")
print(f"Day: {today.day}")
```

```python
# Get current date AND time
import datetime

now = datetime.datetime.now()
print(f"Full: {now}")
print()

# Format it nicely
formatted = now.strftime("%B %d, %Y at %I:%M %p")
print(f"Formatted: {formatted}")
```

```python
# Practical example: Calculate age
import datetime

def calculate_age(birth_year):
    current_year = datetime.date.today().year
    age = current_year - birth_year
    return age

my_age = calculate_age(2000)
print(f"Born in 2000, you are {my_age} years old")
```

```python
# Practical example: Create specific dates
import datetime

# Create a date for Christmas
christmas = datetime.date(2025, 12, 25)
today = datetime.date.today()

days_until = (christmas - today).days
print(f"Days until Christmas: {days_until}")
```

## 4. How to Import - Different Methods

```python
# Method 1: Import the whole module (MOST COMMON)
import random
print(random.randint(1, 10))
print()

# Method 2: Import specific functions
from random import randint, choice
print(randint(1, 10))  # No need for random.
print(choice(["a", "b", "c"]))
print()

# Method 3: Import with a nickname
import datetime as dt
today = dt.date.today()
print(today)
```

## 5. Putting It Together - Larger Example

```python
# Daily fortune generator
import random
import datetime

def generate_fortune():
    fortunes = [
        "Today is your lucky day!",
        "Be cautious with your decisions.",
        "Good news is coming your way.",
        "Take a chance on something new.",
        "Focus on your goals today."
    ]
    
    lucky_numbers = []
    for i in range(3):
        lucky_numbers.append(random.randint(1, 50))
    
    today = datetime.date.today()
    fortune = random.choice(fortunes)
    
    return {
        "date": today.strftime("%B %d, %Y"),
        "fortune": fortune,
        "lucky_numbers": lucky_numbers
    }

# Get today's fortune
result = generate_fortune()
print("=" * 40)
print(f"Date: {result['date']}")
print(f"Fortune: {result['fortune']}")
print(f"Lucky Numbers: {result['lucky_numbers']}")
print("=" * 40)
```

```python
# Password generator
import random

def generate_password(length):
    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%"
    password = ""
    
    for i in range(length):
        password += random.choice(characters)
    
    return password

# Generate 3 passwords
print("Generated passwords:")
for i in range(3):
    pwd = generate_password(10)
    print(f"  {i+1}. {pwd}")
```

## Key Takeaways

**Importing:**
- `import module_name` (most common)
- `from module import function`
- `import module as nickname`

**Random Module - Add randomness:**
- `random.randint(a, b)` - random integer
- `random.choice(list)` - random item from list
- `random.shuffle(list)` - randomize list order

**Datetime Module - Work with dates:**
- `datetime.date.today()` - get today's date
- `datetime.datetime.now()` - get current date and time
- `.strftime()` - format dates nicely

**Why These Matter:**
- You can't generate true randomness without `random`
- Dates are complex - let `datetime` handle it
- Focus on solving problems, not reinventing tools
