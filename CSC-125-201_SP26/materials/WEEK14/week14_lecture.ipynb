## Reading Error Messages

Read from **bottom to top**:
- **What** type of error (KeyError, ValueError, etc.)
- **Where** (line number)
- **Why** it happened (description)

## Debugging Strategy

**Use print statements to:**
- See what values variables have
- Check if conditions are True or False
- Track which parts of code are running
- See what's in dictionaries and lists

**Most common fixes:**
- Clean input: `user_input.strip().lower()`
- Check keys: `if key in dictionary`
- Return from functions: `return result`
- Initialize before loop: `total = 0`
