## 1. You Already Know Functions

You've been using functions all semester:

```python
# Functions you've been using
print("Hello")           # print() is a function
len([1, 2, 3])           # len() is a function
max([5, 2, 8])           # max() is a function
input("Enter name: ")    # input() is a function

# String methods are functions too
text = "hello"
text.upper()             # .upper() is a function
text.split()             # .split() is a function

# List methods
numbers = [1, 2, 3]
numbers.append(4)        # .append() is a function

# Functions take input and do something or return something
```

## 2. Why Create Your Own Functions?

Let's say you need to greet students multiple times:

```python
# Without functions - repetitive code
print("=" * 30)
print("Hello, Alice")
print("Welcome to CS 125")
print("=" * 30)
print()

print("=" * 30)
print("Hello, Bob")
print("Welcome to CS 125")
print("=" * 30)
print()

print("=" * 30)
print("Hello, Charlie")
print("Welcome to CS 125")
print("=" * 30)

# This is tedious. What if we want to change the message?
# We'd have to change it in 3 places
```

## 3. Creating Your First Function

Syntax: `def function_name():`

```python
# Define a function
def greet():
    print("=" * 30)
    print("Hello")
    print("Welcome to CS 125")
    print("=" * 30)

# Call the function
greet()
print()
greet()
print()
greet()

# Write once, use many times
```

```python
# Another example: Calculate area of rectangle
def calculate_area():
    length = 10
    width = 5
    area = length * width
    print(f"Area: {area}")

calculate_area()
calculate_area()

# But this always calculates the same rectangle
# We need to pass in different values
```

## 4. Functions with Parameters

Parameters let you pass data into functions:

```python
# Function with one parameter
def greet(name):
    print("=" * 30)
    print(f"Hello, {name}")
    print("Welcome to CS 125")
    print("=" * 30)

# Call with different arguments
greet("Alice")
print()
greet("Bob")
print()
greet("Charlie")

# name is the parameter (in definition)
# "Alice", "Bob", "Charlie" are arguments (when calling)
```

```python
# Function with multiple parameters
def calculate_area(length, width):
    area = length * width
    print(f"Rectangle {length} x {width}")
    print(f"Area: {area}")
    print()

# Call with different values
calculate_area(10, 5)
calculate_area(7, 3)
calculate_area(15, 8)

# Order matters
calculate_area(10, 5)  # length=10, width=5
calculate_area(5, 10)  # length=5, width=10 (different rectangle)
```

```python
# More examples with parameters
def print_price(item, price):
    print(f"{item}: ${price:.2f}")

print_price("Apple", 1.50)
print_price("Banana", 0.75)
print_price("Orange", 2.00)
print()

def check_passing(name, score):
    print(f"{name}: {score}")
    if score >= 60:
        print("  Status: PASS")
    else:
        print("  Status: FAIL")
    print()

check_passing("Alice", 85)
check_passing("Bob", 55)
check_passing("Charlie", 92)
```

## 5. Return Values

Sometimes you want to get a value back from a function:

```python
# Function that prints
def calculate_area_print(length, width):
    area = length * width
    print(f"Area: {area}")

calculate_area_print(10, 5)
# Can't use the result in calculations

print()

# Function that returns
def calculate_area_return(length, width):
    area = length * width
    return area

# Now we can use the result
room1 = calculate_area_return(10, 5)
room2 = calculate_area_return(8, 6)
total = room1 + room2

print(f"Room 1: {room1}")
print(f"Room 2: {room2}")
print(f"Total area: {total}")
```

```python
# Compare print vs return
def add_print(a, b):
    result = a + b
    print(result)  # Shows the value

def add_return(a, b):
    result = a + b
    return result  # Gives back the value

print("Using print version:")
add_print(5, 3)
# Can't do: total = add_print(5, 3) + 10
print()

print("Using return version:")
total = add_return(5, 3)
print(f"Result: {total}")
doubled = add_return(5, 3) * 2
print(f"Doubled: {doubled}")
```

```python
# More return examples
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# Use the returned value
alice_grade = get_grade(92)
bob_grade = get_grade(78)

print(f"Alice: {alice_grade}")
print(f"Bob: {bob_grade}")
print()

# Can use in conditionals
if get_grade(85) == "B":
    print("Got a B")
```

## 6. Functions with Loops and Data Structures

```python
# Function that works with lists
def calculate_average(numbers):
    total = sum(numbers)
    avg = total / len(numbers)
    return avg

scores1 = [85, 92, 78, 95]
scores2 = [70, 88, 82, 91]

avg1 = calculate_average(scores1)
avg2 = calculate_average(scores2)

print(f"Class 1 average: {avg1:.1f}")
print(f"Class 2 average: {avg2:.1f}")
print()

# Function that finds maximum
def find_max(numbers):
    highest = numbers[0]
    for num in numbers:
        if num > highest:
            highest = num
    return highest

print(f"Highest in class 1: {find_max(scores1)}")
print(f"Highest in class 2: {find_max(scores2)}")
```

```python
# Function that works with dictionaries
def get_contact_info(contacts, name):
    if name in contacts:
        return contacts[name]
    else:
        return "Not found"

phone_book = {
    "Alice": "555-1234",
    "Bob": "555-5678",
    "Charlie": "555-9012"
}

print(get_contact_info(phone_book, "Alice"))
print(get_contact_info(phone_book, "Diana"))
print()

# Function that counts something
def count_passing(grades):
    count = 0
    for grade in grades.values():
        if grade >= 60:
            count += 1
    return count

class_grades = {"Alice": 85, "Bob": 55, "Charlie": 92, "Diana": 78}
passing = count_passing(class_grades)
print(f"Students passing: {passing} out of {len(class_grades)}")
```

## 7. Putting It Together - Practical Example

```python
# Grade analyzer with functions

def calculate_average(scores):
    return sum(scores) / len(scores)

def get_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

def print_report(name, scores):
    avg = calculate_average(scores)
    grade = get_letter_grade(avg)
    
    print(f"\n{name}:")
    print(f"  Scores: {scores}")
    print(f"  Average: {avg:.1f}")
    print(f"  Grade: {grade}")

# Use the functions
students = {
    "Alice": [85, 92, 88, 90],
    "Bob": [78, 82, 75, 80],
    "Charlie": [92, 95, 89, 94]
}

print("=" * 40)
print("GRADE REPORT")
print("=" * 40)

for name, scores in students.items():
    print_report(name, scores)
```

## Key Takeaways

**Function Basics:**
- Define with `def function_name():`
- Call with `function_name()`
- Body must be indented

**Parameters:**
- Pass data in: `def greet(name):`
- Multiple parameters: `def area(length, width):`
- Order matters when calling

**Return Values:**
- Use `return` to send data back
- Can use returned value in expressions
- Without return, function returns `None`

**Why Functions?**
- DRY: Don't Repeat Yourself
- Organize code into reusable pieces
- Easier to test and debug
- Make code more readable
