## Part 1: Basic For Loops

A for loop repeats code once for each item in a list. The loop variable takes on each value in turn.

```python
# DEMO - Run this cell and observe the results.
tasks = ["Review report", "Send invoice", "Update spreadsheet", "Schedule meeting"]

print("Today's tasks:")
for task in tasks:
    print(f"  - {task}")

print(f"\nTotal tasks: {len(tasks)}")
```

## Part 3: The range() Function

`range(stop)` counts from 0. `range(start, stop)` counts from start. `range(start, stop, step)` counts by a custom amount. All three stop before the stop value.

```python
# DEMO - Run this cell and observe the results.
print("Appointment slots:")
for hour in range(9, 13):
    print(f"  {hour}:00 AM - Available")

print()
print("Countdown:")
for n in range(5, 0, -1):
    print(n, end=" ")
print("- Done.")
```

## Part 4: Index-Based Loops and Parallel Lists

Use `range(len(list))` when you need the index number, need to modify a list in place, or need to access multiple lists at the same position.

```python
# DEMO - Run this cell and observe the results.
courses = ["Math", "English", "History"]
credits = [4, 3, 3]

print("Course Load:")
total_credits = 0
for i in range(len(courses)):
    print(f"  {courses[i]}: {credits[i]} credits")
    total_credits = total_credits + credits[i]

print(f"Total: {total_credits} credits")
```

## Part 5: Looping Through Strings

Strings are sequences. A for loop visits each character one at a time, which is useful for counting, filtering, or building new strings.

```python
# DEMO - Run this cell and observe the results.
phone = "(617) 555-0142"

digits_only = ""
for char in phone:
    if char.isdigit():
        digits_only = digits_only + char

print(f"Formatted: {phone}")
print(f"Digits only: {digits_only}")
```
