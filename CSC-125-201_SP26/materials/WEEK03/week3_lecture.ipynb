### String Methods

Strings come with built-in tools called **methods**. You call them with dot notation:
```
string.method()
```

Methods return a **new** string - the original is unchanged.

````python
# Let's start with some messy user input
user_input = "  HELLO world  "
print(f"Original: '{user_input}'")
print(f"Length: {len(user_input)}")
````

````python
# .strip() removes leading and trailing spaces
clean = user_input.strip()
print(f"After strip(): '{clean}'")

# .upper() converts to all uppercase
print(f"After upper(): '{user_input.upper()}'")

# .lower() converts to all lowercase
print(f"After lower(): '{user_input.lower()}'")

# .title() capitalizes the first letter of each word
print(f"After title(): '{user_input.title()}'")
````

````python
# The original string is NOT changed
print(f"Original is still: '{user_input}'")

# You must save the result to use it
cleaned_input = user_input.strip().lower()
print(f"Saved result: '{cleaned_input}'")
````

### Method Chaining

You can apply multiple methods in sequence:

````python
# Method chaining: apply multiple methods in one line
messy_name = "  alice JOHNSON  "

# Chain strip and title together
clean_name = messy_name.strip().title()
print(f"Original: '{messy_name}'")
print(f"Cleaned:  '{clean_name}'")

# This is the same as doing it step by step:
step1 = messy_name.strip()
step2 = step1.title()
print(f"Step by step: '{step2}'")
````

### The .replace() Method

Replace parts of text with something else:

````python
order = "I want a large pizza with EXTRA cheese"
print(f"Original: {order}")

# Replace one word
fixed = order.replace("EXTRA", "extra")
print(f"Fixed: {fixed}")

# Replace a phrase
formal = order.replace("I want", "I would like")
print(f"Formal: {formal}")
````

### String Indexing

Each character in a string has a **position number** (index), starting at **0**:

````python
word = "Python"
print(f"The word: {word}")
print(f"Length: {len(word)}")
print()

# Positive indexing (left to right, starting at 0)
print("Positive indexing:")
print(f"word[0] = '{word[0]}'  # First character")
print(f"word[1] = '{word[1]}'  # Second character")
print(f"word[2] = '{word[2]}'  # Third character")
print(f"word[5] = '{word[5]}'  # Last character")
````

````python
# Negative indexing (right to left, starting at -1)
print("Negative indexing:")
print(f"word[-1] = '{word[-1]}'  # Last character")
print(f"word[-2] = '{word[-2]}'  # Second to last")
print(f"word[-6] = '{word[-6]}'  # First character")
````

### Practical Indexing: Getting Initials

````python
full_name = "Alice Johnson"
print(f"Full name: {full_name}")

# Find the space to locate where last name starts
space_position = full_name.index(" ")
print(f"Space is at position: {space_position}")

# Get initials
first_initial = full_name[0]
last_initial = full_name[space_position + 1]
print(f"Initials: {first_initial}.{last_initial}.")
````

### String Slicing

Slicing extracts **multiple characters** using `[start:end]`.

The end index is **not included** (it stops just before it).

````python
sentence = "Hello World"
print(f"Sentence: {sentence}")
print(f"Length: {len(sentence)}")
print()

# Basic slicing: [start:end] - end is NOT included
print(f"sentence[0:5] = '{sentence[0:5]}'  # Characters 0,1,2,3,4")
print(f"sentence[6:11] = '{sentence[6:11]}'  # Characters 6,7,8,9,10")
````

````python
# Shorthand slicing
print(f"sentence[:5] = '{sentence[:5]}'   # From start to position 5")
print(f"sentence[6:] = '{sentence[6:]}'   # From position 6 to end")
print(f"sentence[:] = '{sentence[:]}'     # The entire string")
````

### Practical Slicing: Email Parser

````python
# Extract username and domain from an email
email = "student@college.edu"
print(f"Email: {email}")

# Find where the @ symbol is
at_position = email.index("@")
print(f"@ is at position: {at_position}")

# Extract parts
username = email[:at_position]
domain = email[at_position + 1:]

print(f"Username: {username}")
print(f"Domain: {domain}")
````

### F-String Formatting

F-strings let you insert variables directly into strings with `f"..."`:

````python
# Traditional way: concatenation (messy)
name = "Alice"
age = 22
old_way = "Hello, my name is " + name + " and I am " + str(age) + " years old."
print("Old way:", old_way)

# Modern way: f-strings (clean)
f_way = f"Hello, my name is {name} and I am {age} years old."
print("F-string:", f_way)
````

````python
# F-strings can contain expressions and method calls
name = "bob"
age = 25
city = "Boston"

print(f"Hello {name.title()}, welcome to {city}.")
print(f"Next year you will be {age + 1}.")
print(f"Your name in caps: {name.upper()}")
````

````python
# Formatting decimal numbers with :.2f
price = 19.99
tax_rate = 0.08
tax = price * tax_rate
total = price + tax

print(f"Price: ${price}")
print(f"Tax: ${tax:.2f}")
print(f"Total: ${total:.2f}")
````

### Understanding Error Messages

Errors are normal. The key is learning to **read** them.

Every error message tells you:
1. **Error type** (last line)
2. **Location** (the arrow `--->`)
3. **Description** (what went wrong)

### Error Type 1: IndexError

Accessing a position that does not exist:

````python
text = "Hello"
print(f"Text: {text}")
print(f"Length: {len(text)}")
print(f"Valid indexes: 0, 1, 2, 3, 4")

# This works:
print(f"text[4] = '{text[4]}'")

# Uncomment to see an IndexError:
# print(f"text[10] = '{text[10]}'")
````

### Error Type 2: AttributeError

Using a method name that does not exist (usually a capitalization issue):

````python
name = "alice"
print(f"Correct: {name.upper()}")

# Uncomment to see an AttributeError:
# print(f"Wrong: {name.Upper()}")  # Capital U - does not exist
````

### Error Type 3: NameError

Using a variable that has not been defined (usually a typo):

````python
color = "blue"
print(f"My favorite color is {color}")

# Uncomment to see a NameError:
# print(f"I also like {colour}")  # Typo: 'colour' vs 'color'
````

### Debugging Strategy: Print and Check

When something goes wrong, add `print()` statements to inspect your data:

````python
# Debugging example: check each step
email = "student@college.edu"
print(f"Working with: {email}")
print(f"Length: {len(email)}")
print(f"Type: {type(email)}")

at_pos = email.index("@")
print(f"@ at position: {at_pos}")

username = email[:at_pos]
domain = email[at_pos + 1:]
print(f"Username: '{username}'")
print(f"Domain: '{domain}'")
````
