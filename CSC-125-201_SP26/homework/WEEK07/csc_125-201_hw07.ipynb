## Part 1: Advanced Slicing

The full slicing syntax is `list[start:end:step]`. The step controls how many items to skip. A negative step walks backward, and `[::-1]` reverses a list.

```python
# DEMO - Run this cell and observe the results.
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Every other month starting from January
alternating = months[::2]
print(f"Every other: {alternating}")

# Last 4 months in reverse
last_four_rev = months[-1:-5:-1]
print(f"Last 4 reversed: {last_four_rev}")

# Reverse the whole list
backward = months[::-1]
print(f"Dec to Jan: {backward}")
```

## Part 2: Sorting

`.sort()` modifies a list in place and returns `None`. `sorted()` returns a new sorted list and leaves the original unchanged. Both accept `reverse=True` for descending order.

```python
# DEMO - Run this cell to see the difference.
wait_times = [12, 45, 8, 23, 31]

# sorted() returns a new list
shortest_first = sorted(wait_times)
print(f"Sorted copy: {shortest_first}")
print(f"Original: {wait_times}")

# .sort() changes the original, returns None
result = wait_times.sort()
print(f"After .sort(): {wait_times}")
print(f".sort() returned: {result}")
```

## Part 3: Searching and Counting

Use `in` to check if a value exists, `.index()` to find its position, and `.count()` to tally occurrences. Always check with `in` before calling `.index()` to avoid errors.

```python
# DEMO - Run this cell.
genres = ["rock", "jazz", "pop", "rock", "blues", "rock"]

print(f"'jazz' in list: {'jazz' in genres}")
print(f"'jazz' position: {genres.index('jazz')}")
print(f"'rock' count: {genres.count('rock')}")
```

## Part 4: Building and Copying Lists

`.insert(i, x)` adds at a specific position. `.extend(other)` merges another list. `.copy()` or `[:]` creates an independent copy.

```python
# DEMO - Run this cell.
tools = ["hammer", "wrench"]
more_tools = ["pliers", "saw"]

tools.insert(0, "drill")
print(f"After insert: {tools}")

tools.extend(more_tools)
print(f"After extend: {tools}")

backup = tools.copy()
backup.append("level")
print(f"Backup: {backup}")
print(f"Original: {tools}")
```

## Part 5: String-List Conversion

`.split(separator)` breaks a string into a list. `separator.join(list)` combines a list of strings back into one.

```python
# DEMO - Run this cell.
ingredients = "flour, sugar, butter, eggs"
parts = ingredients.split(", ")
print(f"List: {parts}")

rejoined = " + ".join(parts)
print(f"Rejoined: {rejoined}")
```
