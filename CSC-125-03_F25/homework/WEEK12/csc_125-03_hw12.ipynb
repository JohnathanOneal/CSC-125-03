## Part 4: Functions with Lists

Create functions that process lists.

```python
# Example: Function that finds sum of a list
def calculate_sum(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

values = [10, 20, 30, 40]
result = calculate_sum(values)
print(f"Sum: {result}")
```

## Part 5: Functions that Return Lists

Create functions that build and return new lists.

```python
# Example: Function that creates a new list
def add_ten(numbers):
    result = []
    for num in numbers:
        result.append(num + 10)
    return result

original = [5, 15, 25]
new_list = add_ten(original)
print(f"Original: {original}")
print(f"New: {new_list}")
```

## Part 7: Combining Functions

Use multiple functions together to solve problems.

```python
# Example: Multiple functions working together
def double(x):
    return x * 2

def add_five(x):
    return x + 5

def process_number(x):
    doubled = double(x)
    result = add_five(doubled)
    return result

print(process_number(10))  # (10 * 2) + 5 = 25
```
