### What is a Variable?

A variable stores data in your program. Think of it as a **labeled box**:
- The box = the computer's memory space
- The label = the variable name
- What's inside = the value

We create variables with the `=` assignment operator:
```
variable_name = value
```

````python
hero_name = "Batman"
print(hero_name)
````

````python
# Variables let you reuse and update values easily
hero_name = "Spider-Man"
city = "New York"

print("The amazing", hero_name, "protects", city)
print(hero_name, "swings through the streets of", city)
print("Citizens of", city, "love", hero_name)
````

````python
# Variables can be updated
favorite_food = "pizza"
print("Favorite food:", favorite_food)

favorite_food = "tacos"
print("Favorite food:", favorite_food)
````

### The Four Data Types

| Type | Stores | Example | Python Name |
|------|--------|---------|-------------|
| String | Text | `"Hello"` | `str` |
| Integer | Whole numbers | `42` | `int` |
| Float | Decimal numbers | `3.14` | `float` |
| Boolean | True/False | `True` | `bool` |

````python
# Strings - text in quotes
name = "Alice"
favorite_color = 'blue'
number_as_text = "123"  # still a string because of the quotes

print("Name:", name)
print("Color:", favorite_color)
print("Type of name:", type(name))
````

````python
# Integers - whole numbers
age = 25
temperature = -5
score = 0

print("Age:", age)
print("Type of age:", type(age))
print("Age in 5 years:", age + 5)
````

````python
# Floats - decimal numbers
pi = 3.14159
price = 19.99
exact_zero = 0.0  # float because of the decimal

print("Pi:", pi)
print("Type of pi:", type(pi))
print("Circumference:", 2 * pi * 5)
````

````python
# Booleans - True or False
is_student = True
has_license = False

age = 20
can_vote = age >= 18  # comparison produces a boolean

print("Is student:", is_student)
print("Can vote:", can_vote)
print("Type:", type(is_student))
````

````python
# Summary: all four types side by side
my_string = "Hello World"
my_integer = 42
my_float = 3.14
my_boolean = True

print("String:", my_string, "- Type:", type(my_string))
print("Integer:", my_integer, "- Type:", type(my_integer))
print("Float:", my_float, "- Type:", type(my_float))
print("Boolean:", my_boolean, "- Type:", type(my_boolean))
````

### Variable Naming Rules

**Rules:**
1. Must start with a letter or underscore
2. Cannot start with a number
3. Only letters, numbers, and underscores
4. No spaces, no dashes
5. Cannot use Python keywords
6. Case sensitive (`Name` and `name` are different)

````python
# Valid names
user_name = "John"
age2 = 25
is_valid = True
_private_var = "secret"
CONSTANT_VALUE = 100

print(user_name, age2, is_valid)
````

````python
# Invalid names - these cause errors (uncomment one at a time to see)

# 2age = 25          # cannot start with number
# user name = "John" # no spaces
# user-name = "Jane" # no dashes
# print = "Hello"    # Python keyword
````

**Best practices:**
- Descriptive names: `student_grade` not `sg`
- snake_case: `first_name` not `firstName`
- Lowercase for variables, UPPERCASE for constants
- Booleans: prefix with `is_` or `has_`

````python
# Well-named variables
student_name = "Alice"
final_grade = 87.5
is_passing = True
MAX_STUDENTS = 30

print(f"Student: {student_name}")
print(f"Grade: {final_grade}")
print(f"Passing: {is_passing}")
print(f"Class limit: {MAX_STUDENTS}")
````

### Type Conversion

Convert between types with `int()`, `float()`, `str()`

````python
# Converting a string to a number
text_number = "25"
real_number = int(text_number)

print("Text:", text_number, "- Type:", type(text_number))
print("Number:", real_number, "- Type:", type(real_number))
print("Number + 5 =", real_number + 5)
````

````python
# More conversions
number = 42
decimal = 3.14

print(f"Number as string: {str(number)} ({type(str(number))})")
print(f"Decimal as integer: {int(decimal)} ({type(int(decimal))})")
print(f"Number as float: {float(number)} ({type(float(number))})")
````

### The input() Function

`input()` asks the user for information. It **always returns a string**.

````python
user_name = input("What's your name? ")
print("Hello,", user_name)
print("Type:", type(user_name))
````

````python
# For numbers, you must convert the input
age_text = input("How old are you? ")
age_number = int(age_text)
print(f"Next year you'll be {age_number + 1}")
````

````python
# Shortcut: convert in one line
age = int(input("Enter your age: "))
height = float(input("Enter your height in feet: "))

print("Age:", age, "- Type:", type(age))
print("Height:", height, "- Type:", type(height))
````

### Basic Arithmetic with Variables

````python
x = 10
y = 3

print(f"{x} + {y} = {x + y}")
print(f"{x} - {y} = {x - y}")
print(f"{x} * {y} = {x * y}")
print(f"{x} / {y} = {x / y}")
````
