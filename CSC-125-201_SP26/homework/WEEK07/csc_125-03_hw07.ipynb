## Part 2: List Slicing

Use slicing to extract portions of lists.

```python
# Example: Days of the week
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Different slices
weekdays = days[0:5]  # First 5 days
weekend = days[5:]    # Last 2 days
first_three = days[:3]

print(f"Weekdays: {weekdays}")
print(f"Weekend: {weekend}")
print(f"First three days: {first_three}")
```

## Part 5: Tuples

```python
# Example: Coordinates and unpacking
point = (10, 20)
print(f"Point: {point}")

# Unpacking
x, y = point
print(f"x = {x}, y = {y}")

# Swapping values
a = 5
b = 10
print(f"Before swap: a={a}, b={b}")
a, b = b, a
print(f"After swap: a={a}, b={b}")
```

## Part 7: Copying Lists

Understand the difference between references and copies.

```python
# Example: Reference vs Copy
original = [1, 2, 3]
reference = original  # This is NOT a copy
copy = original[:]    # This IS a copy

print(f"Original: {original}")

# Modify the copy
copy.append(4)
print(f"After modifying copy: {original} vs {copy}")

# Modify the reference
reference.append(5)
print(f"After modifying reference: {original} vs {reference}")
```
