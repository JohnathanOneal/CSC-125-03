## Part 3: Operator Precedence

Python follows the mathematical order of operations (PEMDAS). Parentheses let you control the order.

```python
# DEMO - Run this cell to see how precedence changes the result
result_no_parens = 2 + 3 * 4
result_with_parens = (2 + 3) * 4

print(f"2 + 3 * 4   = {result_no_parens}")
print(f"(2 + 3) * 4 = {result_with_parens}")
print()
print("Without parentheses: multiplication happens first (3 * 4 = 12, then 2 + 12 = 14)")
print("With parentheses: addition happens first (2 + 3 = 5, then 5 * 4 = 20)")
```

## Part 5: Assignment Operators

Assignment operators (`+=`, `-=`, `*=`, `/=`) let you update a variable based on its current value.

```python
# DEMO - Run this cell to see assignment operators in action
score = 0
print(f"Starting score: {score}")

score += 100    # same as score = score + 100
print(f"After += 100:   {score}")

score -= 25     # same as score = score - 25
print(f"After -= 25:    {score}")

score *= 2      # same as score = score * 2
print(f"After *= 2:     {score}")
```
