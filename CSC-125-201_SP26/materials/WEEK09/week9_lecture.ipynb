### The Limitation of for Loops

A for loop is great when you have a list to process or a known number of repetitions. But what about situations where you do not know in advance how many times to repeat?

- Asking a user for a valid password (they might get it right on the first try, or the tenth)
- A game that keeps running until the player loses
- Reading data until you reach a special "stop" value

In all of these, you repeat **until a condition changes**, not for a fixed number of times.

---

### Basic While Loop Syntax

A while loop checks a condition **before** each repetition. If the condition is True, the body runs. Then it checks again. When the condition becomes False, the loop stops.

```python
while condition:
    # body: runs as long as condition is True
```

The condition is checked **only at the top**, not continuously during the body. Think of a gatekeeper: they check your ticket at the door, not while you are inside.

````python
# Basic counting with a while loop
count = 1
while count <= 5:
    print(f"Count is: {count}")
    count = count + 1

print(f"\nLoop ended. count is now: {count}")
````

### Hand-Tracing a While Loop

The best way to understand a loop is to trace it by hand. Make a table with one column per variable and one row per iteration. Cross out old values instead of erasing them, so you can see the progression.

Let's trace the loop above:

| Iteration | `count` (start of loop) | `count <= 5` | Action | `count` (end of body) |
|-----------|------------------------|--------------|--------|-----------------------|
| 1 | 1 | True | print 1 | 2 |
| 2 | 2 | True | print 2 | 3 |
| 3 | 3 | True | print 3 | 4 |
| 4 | 4 | True | print 4 | 5 |
| 5 | 5 | True | print 5 | 6 |
| 6 | 6 | False | skip body | - |

Notice: the condition is checked at the **start** of each iteration, and `count` ends at 6 (one past the stopping point).

````python
# Countdown example
seconds = 5
while seconds > 0:
    print(f"{seconds}...")
    seconds = seconds - 1

print("Go.")
````

````python
# Doubling until a threshold
# How many times does a penny need to double to reach $100?
amount = 0.01
doublings = 0

while amount < 100:
    amount = amount * 2
    doublings = doublings + 1

print(f"After {doublings} doublings: ${amount:.2f}")
````

### The "Keep Going While" Trap

A very common mistake: you think "stop when count reaches 5" and write `while count == 5`. But the condition describes when to **keep going**, not when to stop.

| What you want | Wrong condition | Right condition |
|---------------|---------------|----------------|
| Stop at 5 | `while count == 5` | `while count < 5` |
| Stop when empty | `while len(items) == 0` | `while len(items) > 0` |
| Stop when user says quit | `while answer == "quit"` | `while answer != "quit"` |

**Tip:** read the while line out loud as "keep going while..." to check if the logic makes sense.

### Infinite Loops

If nothing inside the loop changes the condition toward False, the loop runs forever. This is called an infinite loop. In Kaggle, the cell will show a spinning indicator that never stops.

**How to stop an infinite loop in Kaggle:** click the stop button next to the running cell, or go to Run > Interrupt Execution.

The most common cause: forgetting to update the loop variable.

````python
# CAUTION: This would be an infinite loop if we ran it.
# The condition never becomes False because count never changes.
#
# count = 1
# while count <= 5:
#     print(count)
#     # Missing: count = count + 1

# Fixed version:
count = 1
while count <= 5:
    print(count)
    count = count + 1  # This line prevents the infinite loop
````

### For Loop vs While Loop

When should you use which?

| Situation | Use |
|-----------|-----|
| Process every item in a list | `for` |
| Repeat a known number of times | `for` with `range()` |
| Repeat until a condition changes | `while` |
| Keep asking for user input | `while` |
| Run a game until it ends | `while` |

**Rule of thumb:** if you have a list or a known count, use `for`. If you are waiting for something, use `while`.

````python
# Same task, two approaches:

# for loop: print numbers 1 to 5
print("Using for:")
for i in range(1, 6):
    print(i, end=" ")

print()

# while loop: print numbers 1 to 5
print("\nUsing while:")
n = 1
while n <= 5:
    print(n, end=" ")
    n = n + 1

print()
print("\nFor this task, 'for' is simpler. Use 'while' when you can't predict the count.")
````

### break: The Emergency Exit

The `break` statement immediately ends the loop, no matter what the condition says. Python jumps to the first line after the loop.

Use `break` when:
- You found what you were searching for
- The user types a special "quit" command
- A condition makes continuing pointless

````python
# Search for a value in a list
numbers = [4, 8, 15, 16, 23, 42]
target = 16

found = False
index = 0
while index < len(numbers):
    if numbers[index] == target:
        found = True
        break  # No need to keep searching
    index = index + 1

if found:
    print(f"Found {target} at index {index}")
else:
    print(f"{target} not found")
````

````python
# A menu that runs until the user types "quit"
# (Using a predefined list to simulate user input in Kaggle)
simulated_inputs = ["hello", "how are you", "quit", "this won't run"]

i = 0
while True:  # Runs forever unless we break
    command = simulated_inputs[i]
    i = i + 1

    if command == "quit":
        print("Goodbye.")
        break

    print(f"You said: {command}")

print("Program ended.")
````

### continue: Skip This Lap

Where `break` ends the entire loop, `continue` skips only the **rest of the current iteration** and jumps back to the condition check.

Think of running laps on a track. You trip and decide to skip the rest of this lap, but you do not leave the track. You walk back to the start line and run the next lap normally.

````python
# Process orders, but skip any that are zero or negative
orders = [25.00, 0, 18.50, -5.00, 42.75, 0, 31.00]

total = 0
valid_count = 0
index = 0

while index < len(orders):
    amount = orders[index]
    index = index + 1  # Always increment BEFORE continue

    if amount <= 0:
        print(f"  Skipping invalid amount: {amount}")
        continue  # Skip the rest, go back to condition

    total = total + amount
    valid_count = valid_count + 1
    print(f"  Processed: ${amount:.2f}")

print(f"\nTotal from {valid_count} valid orders: ${total:.2f}")
````

**Important:** when using `continue` in a while loop, make sure you update the loop variable **before** the continue statement. Otherwise you create an infinite loop, because `continue` skips the update line.

### Input Validation

One of the most practical while loop patterns. Keep asking the user for input until they give something valid.

Since we cannot use `input()` in Kaggle the same way as a desktop Python environment, we will simulate user input with a list. The pattern is the same.

````python
# Simulating input validation
# Imagine the user is asked to enter a number between 1 and 10
user_attempts = ["abc", "-3", "15", "7"]  # Simulated inputs

attempt = 0
valid = False

while not valid:
    response = user_attempts[attempt]
    attempt = attempt + 1
    print(f"User entered: '{response}'")

    # Check if it is a number
    if not response.isdigit() and not (response.startswith("-") and response[1:].isdigit()):
        print("  Not a valid number. Try again.")
        continue

    number = int(response)

    # Check if it is in range
    if number < 1 or number > 10:
        print(f"  {number} is not between 1 and 10. Try again.")
        continue

    valid = True

print(f"\nAccepted: {number}")
````

### Sentinel Values

A sentinel value is a special value that signals "stop." Common examples: -1, 0, "quit", or an empty string. The loop checks for this value each iteration.

````python
# Add up numbers until the user enters -1 (sentinel value)
user_numbers = [10, 25, 8, 42, -1]  # Simulated inputs

total = 0
index = 0
number = user_numbers[index]

while number != -1:
    total = total + number
    print(f"  Added {number}, total is now {total}")
    index = index + 1
    number = user_numbers[index]

print(f"\nFinal total: {total}")
````

## Advanced Debugging

Debugging is not about guessing. It is a **systematic process**. Most beginners debug by randomly changing things until the error disappears. That rarely works and sometimes introduces new bugs.

### The 5-Step Debugging Process

1. **Reproduce** - Can you make the bug happen again? What input triggers it?
2. **Read** - Read the error message bottom-up. What type of error? What line?
3. **Hypothesize** - Based on the error and the code, what do you think went wrong?
4. **Test** - Add a `print()` to check your hypothesis. Is the variable what you expected?
5. **Fix and verify** - Make one change, run again. Did it fix the problem without creating a new one?

### Common Error Types

| Error Type | What it means | Common cause |
|------------|--------------|---------------|
| `SyntaxError` | Python cannot parse the line | Missing colon, unmatched parenthesis, bad indentation |
| `NameError` | A variable or function name is not defined | Typo in variable name, used before assignment |
| `TypeError` | Wrong data type for an operation | Adding a string to an integer |
| `ValueError` | Right type but wrong value | `int("abc")` |
| `IndexError` | Index is out of range | Accessing `list[5]` when list has 5 items (max index is 4) |
| `ZeroDivisionError` | Dividing by zero | Forgot to check if denominator is 0 |

````python
# Let's see some errors in action

# NameError: using a variable that does not exist
try:
    print(total_score)
except NameError as e:
    print(f"NameError: {e}")

# TypeError: adding incompatible types
try:
    result = "Age: " + 25
except TypeError as e:
    print(f"TypeError: {e}")

# ValueError: invalid conversion
try:
    number = int("hello")
except ValueError as e:
    print(f"ValueError: {e}")

# IndexError: index out of range
try:
    colors = ["red", "blue", "green"]
    print(colors[5])
except IndexError as e:
    print(f"IndexError: {e}")

# ZeroDivisionError
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"ZeroDivisionError: {e}")
````

### Reading Error Messages

Error messages (tracebacks) are read **bottom-up**:
1. The **last line** tells you the error type and description
2. The line above shows the **exact code** that failed
3. Above that is the **file and line number**

Let's practice reading a traceback.

````python
# Run this cell to see a real traceback, then read it bottom-up.
# Step 1: What is the error type?
# Step 2: What line caused it?
# Step 3: What does the error message say?

scores = [85, 92, 78]
total = 0
count = 0

while count <= len(scores):  # Bug: should be < not <=
    total = total + scores[count]
    count = count + 1

print(f"Average: {total / len(scores)}")
````

**Walking through the debugging process:**

1. **Reproduce:** Run the cell. It crashes every time.
2. **Read:** The last line says `IndexError: list index out of range`. The failing line is `total = total + scores[count]`.
3. **Hypothesize:** `count` must be reaching an index that does not exist. The list has 3 items (indices 0, 1, 2), so `count` is probably reaching 3.
4. **Test:** We can add a print statement to verify.

````python
# Step 4: Add a print to test our hypothesis
scores = [85, 92, 78]
total = 0
count = 0

while count <= len(scores):  # len(scores) is 3
    print(f"  count = {count}, len(scores) = {len(scores)}")
    if count < len(scores):  # Safety check to avoid the crash
        total = total + scores[count]
    count = count + 1

# We can see: count goes 0, 1, 2, 3. Index 3 does not exist.
# The condition should be < not <=
````

````python
# Step 5: Fix and verify
scores = [85, 92, 78]
total = 0
count = 0

while count < len(scores):  # Fixed: < instead of <=
    total = total + scores[count]
    count = count + 1

print(f"Average: {total / len(scores):.1f}")  # Should be 85.0
````

### try/except: Catching Errors Gracefully

Some errors are predictable. A user might type letters when you need a number, or a list might contain bad data. Instead of letting the program crash, `try/except` catches the error and lets you handle it.

```python
try:
    # code that might fail
except ErrorType:
    # what to do if it fails
```

If nothing goes wrong, the `except` block is skipped entirely.

````python
# Basic try/except
values = ["10", "abc", "25", "0", "15"]

for val in values:
    try:
        number = int(val)
        print(f"  Converted '{val}' to {number}")
    except ValueError:
        print(f"  Could not convert '{val}' - skipping")
````

````python
# Catching multiple error types
values = ["10", "abc", "0", "25"]

for val in values:
    try:
        number = int(val)
        result = 100 / number
        print(f"  100 / {number} = {result:.1f}")
    except ValueError:
        print(f"  '{val}' is not a valid number")
    except ZeroDivisionError:
        print(f"  Cannot divide by zero")
````

````python
# try/except with while loop: robust input processing
raw_scores = ["85", "ninety", "78", "", "92", "abc", "95"]

valid_scores = []
skipped = 0
index = 0

while index < len(raw_scores):
    try:
        score = int(raw_scores[index])
        valid_scores.append(score)
    except ValueError:
        print(f"  Skipped invalid entry: '{raw_scores[index]}'")
        skipped = skipped + 1
    index = index + 1

print(f"\nValid scores: {valid_scores}")
print(f"Skipped: {skipped}")
if len(valid_scores) > 0:
    total = 0
    for s in valid_scores:
        total = total + s
    print(f"Average of valid: {total / len(valid_scores):.1f}")
````

### Catching All Exceptions

You can use `Exception` to catch any error type. This is useful for debugging, but in production code you should catch specific errors so you know exactly what went wrong.

````python
# Catching any exception (useful for debugging)
test_values = ["10", "abc", "0"]

for val in test_values:
    try:
        number = int(val)
        result = 100 / number
        print(f"  100 / {number} = {result:.1f}")
    except Exception as e:
        print(f"  Error with '{val}': {type(e).__name__}: {e}")
````

### Print Debugging: Checking Your Assumptions

Logic errors do not produce tracebacks. The code runs, but the answer is wrong. The key question is: **where does reality diverge from your expectation?**

Add `print()` statements at key points to see what your variables actually contain.

````python
# This code is supposed to count down from 10 to 1.
# But it has a logic error. Let's find it with print debugging.

count = 10
while count >= 1:
    count = count - 1  # Bug: decrement before print
    print(count)

# Expected: 10, 9, 8, ... 1
# Actual: 9, 8, 7, ... 0
# The issue: we subtract BEFORE printing.
````

````python
# Fixed version: print THEN decrement
count = 10
while count >= 1:
    print(count)
    count = count - 1

print("Go.")
````

### Complete Example: Number Guessing Game

This example combines while loops, break, try/except, and input validation into a complete program.

````python
# Number Guessing Game
# (Simulating user input since we're in Kaggle)
import random

secret = 42  # In a real game: random.randint(1, 100)
guesses = [50, 25, 40, 45, 42]  # Simulated user guesses
max_attempts = 7

print("=== Number Guessing Game ===")
print("I'm thinking of a number between 1 and 100.")
print(f"You have {max_attempts} attempts.")
print()

attempt = 0
won = False

while attempt < max_attempts and attempt < len(guesses):
    guess = guesses[attempt]
    attempt = attempt + 1
    print(f"Attempt {attempt}: You guessed {guess}")

    if guess == secret:
        won = True
        break
    elif guess < secret:
        print("  Too low.")
    else:
        print("  Too high.")

    remaining = max_attempts - attempt
    print(f"  {remaining} attempts remaining.")
    print()

if won:
    print(f"\nYou got it in {attempt} attempts.")
else:
    print(f"\nOut of attempts. The number was {secret}.")
````
