### Arithmetic Operators

Python has **7 arithmetic operators**. These are the building blocks for all math in your programs.

```python
# All 7 arithmetic operators
a = 17
b = 5

print(f"Working with a = {a} and b = {b}")
print()

print(f"Addition:         {a} + {b}  = {a + b}")
print(f"Subtraction:      {a} - {b}  = {a - b}")
print(f"Multiplication:   {a} * {b}  = {a * b}")
print(f"Division:         {a} / {b}  = {a / b}")
print(f"Floor Division:   {a} // {b} = {a // b}")
print(f"Modulo:           {a} % {b}  = {a % b}")
print(f"Exponentiation:   {a} ** {b} = {a ** b}")
```

```python
# Division types compared: / vs //
print("Regular division (/) always returns a float:")
print(f"17 / 5  = {17 / 5}")
print(f"10 / 5  = {10 / 5}")   # Even when it divides evenly
print(f"10 / 3  = {10 / 3}")
print()

print("Floor division (//) rounds down to a whole number:")
print(f"17 // 5 = {17 // 5}")
print(f"10 // 5 = {10 // 5}")
print(f"10 // 3 = {10 // 3}")
```

```python
# Practical use: Time conversion with // and %
total_minutes = 135
print(f"Total minutes: {total_minutes}")

hours = total_minutes // 60       # How many full hours?
minutes = total_minutes % 60      # How many minutes left over?

print(f"That is {hours} hours and {minutes} minutes")
```

```python
# Modulo for even/odd checking
# Any number % 2 gives either 0 (even) or 1 (odd)
number = 24
remainder = number % 2
is_even = number % 2 == 0

print(f"{number} % 2 = {remainder}")
print(f"Is {number} even? {is_even}")

number = 37
remainder = number % 2
is_even = number % 2 == 0

print(f"{number} % 2 = {remainder}")
print(f"Is {number} even? {is_even}")
```

### Operator Precedence

Python follows **PEMDAS** (Parentheses, Exponents, Multiplication/Division, Addition/Subtraction) just like math class.

```python
# PEMDAS in action
# Without parentheses: multiplication happens first
result1 = 2 + 3 * 4
print(f"2 + 3 * 4 = {result1}")
print("Step by step: 3 * 4 = 12, then 2 + 12 = 14")
print()

# With parentheses: addition happens first
result2 = (2 + 3) * 4
print(f"(2 + 3) * 4 = {result2}")
print("Step by step: (2 + 3) = 5, then 5 * 4 = 20")
```

```python
# Complex expression traced step by step
result = 10 + 2 * 3 ** 2 - 4 / 2
print(f"10 + 2 * 3 ** 2 - 4 / 2 = {result}")
print()
print("Step by step:")
print("  1. Exponents first:      3 ** 2 = 9")
print("  2. Multiplication:       2 * 9  = 18")
print("  3. Division:             4 / 2  = 2.0")
print("  4. Addition/Subtraction: 10 + 18 - 2.0 = 26.0")
```

```python
# Practical example: calculating an average
# We need parentheses to add BEFORE dividing
temp1 = 72
temp2 = 75
temp3 = 68

# Wrong: only temp3 gets divided
wrong_avg = temp1 + temp2 + temp3 / 3
print(f"Wrong average: {wrong_avg}")
print(f"  (72 + 75 + 68/3 = 72 + 75 + 22.67 = 169.67)")
print()

# Right: all three are added first, then divided
correct_avg = (temp1 + temp2 + temp3) / 3
print(f"Correct average: {correct_avg:.1f}")
print(f"  ((72 + 75 + 68) / 3 = 215 / 3 = 71.7)")
```

### Comparison Operators

Comparison operators compare two values and return `True` or `False`. There are **6 comparison operators**.

```python
# All 6 comparison operators
age = 22

print(f"age = {age}")
print()

print(f"age == 22:  {age == 22}")   # Equal to
print(f"age != 22:  {age != 22}")   # Not equal to
print(f"age > 18:   {age > 18}")    # Greater than
print(f"age < 18:   {age < 18}")    # Less than
print(f"age >= 22:  {age >= 22}")   # Greater than or equal to
print(f"age <= 30:  {age <= 30}")   # Less than or equal to
```

```python
# Comparing strings: case matters
name1 = "Alice"
name2 = "alice"

print(f"'{name1}' == '{name2}': {name1 == name2}")
print("Python sees uppercase and lowercase as different characters.")
print()

# Fix: convert both to the same case before comparing
print(f"'{name1}'.lower() == '{name2}': {name1.lower() == name2}")
```

```python
# Storing comparison results in boolean variables
# This is useful for making your code more readable
age = 22
score = 85
balance = 150.75

can_vote = age >= 18
is_passing = score >= 70
is_positive_balance = balance > 0

print(f"Age: {age} -> Can vote: {can_vote}")
print(f"Score: {score} -> Passing: {is_passing}")
print(f"Balance: ${balance:.2f} -> Positive balance: {is_positive_balance}")
```

### Assignment Operators

Assignment operators are shortcuts for updating a variable's value. Instead of writing `x = x + 5`, you can write `x += 5`.

```python
# Assignment operators: shortcuts for updating variables
score = 100
print(f"Starting score: {score}")

score += 25    # same as score = score + 25
print(f"After += 25:  {score}")

score -= 10    # same as score = score - 10
print(f"After -= 10:  {score}")

score *= 2     # same as score = score * 2
print(f"After *= 2:   {score}")

score /= 5     # same as score = score / 5
print(f"After /= 5:   {score}")
```

```python
# Practical example: shopping cart accumulator
cart_total = 0.00
print(f"Empty cart: ${cart_total:.2f}")

cart_total += 12.99   # Added a book
print(f"Added book ($12.99): ${cart_total:.2f}")

cart_total += 5.49    # Added a pen set
print(f"Added pen set ($5.49): ${cart_total:.2f}")

cart_total += 24.95   # Added a backpack
print(f"Added backpack ($24.95): ${cart_total:.2f}")

print(f"\nCart total: ${cart_total:.2f}")
```
