## Part 3: if-elif-else Chains

When there are more than two outcomes, use `elif`. Python checks conditions top to bottom and runs only the first `True` block.

```python
# DEMO - Run this cell to see an elif chain.
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} -> Grade: {grade}")
```
