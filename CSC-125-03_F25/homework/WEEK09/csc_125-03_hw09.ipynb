## Part 2: Input Validation

Use a while loop to make sure users enter valid data.

```python
# Example: Ask until a number 1–10 is entered
num = input("Enter a number (1–10): ")
while not num.isdigit() or not (1 <= int(num) <= 10):
    print("Invalid input. Try again.")
    num = input("Enter a number (1–10): ")
print(f"You entered {num}!")
```

## Part 3: Break and Continue

Use `break` to exit a loop early and `continue` to skip an iteration.

```python
# Example: Demonstrating break and continue
n = 0
while n < 10:
    n += 1
    if n == 5:
        continue  # Skip 5
    if n == 8:
        break     # Stop at 8
    print(n)
print("Loop ended.")
```

## Part 4: Try/Except Basics

Use `try` and `except` to handle predictable errors and prevent crashes.

```python
# Example: Handle invalid input
try:
    age = int(input("Enter your age: "))
    print(f"Next year you will be {age + 1}.")
except ValueError:
    print("Please enter a number, not text.")
```

## Part 5: Try/Except with Loops

Combine loops with error handling to safely process repeated input.

```python
# Example: Repeat input until valid number is given
while True:
    try:
        number = int(input("Enter an integer: "))
        print(f"Thanks! You entered {number}.")
        break
    except ValueError:
        print("That was not an integer. Try again.")
```
