## 1. Introduction: Why Do We Need Loops?

So far, when we wanted to process multiple items, we had to write repetitive code:

```python
scores = [85, 92, 78, 95, 88]
print(scores[0])
print(scores[1])
print(scores[2])
print(scores[3])
print(scores[4])
```

This is tedious, error-prone, and won't work if we don't know how many items we have in advance!

**For loops** let us repeat actions automatically for each item in a sequence. They're one of the most powerful tools in programming.

## 2. Basic For Loop - Iterating Through Lists

The most common use of for loops is processing each item in a list.

### Basic Syntax
```python
for item in list_name:
    # code to execute for each item
```

**Important:** Notice the colon `:` and the indentation (4 spaces) - just like with if statements!

````python
# Simple for loop through a list
fruits = ["apple", "banana", "cherry", "date"]

print("Fruits in my list:")
for fruit in fruits:
    print(fruit)

print("\nDone printing fruits!")
````

````python
# Processing numbers in a list
scores = [85, 92, 78, 95, 88]

print("Test Scores:")
for score in scores:
    print(f"Score: {score}")
    
    if score >= 90:
        print("  Excellent!")
    elif score >= 80:
        print("  Good job!")
    else:
        print("  Keep working!")
    print()  # Blank line for readability
````

### Building New Lists with For Loops

````python
# Start with empty list, build it up in the loop
prices = [10.99, 25.50, 15.75, 8.25]

# Calculate sales tax for each item
prices_with_tax = []
tax_rate = 0.08

for price in prices:
    price_with_tax = price * (1 + tax_rate)
    prices_with_tax.append(price_with_tax)

print(f"Original prices: {prices}")
print(f"Prices with tax: {prices_with_tax}")
````

````python
# Filtering items from a list
numbers = [3, 8, 15, 6, 22, 9, 18, 5]

# Keep only numbers greater than 10
large_numbers = []

for num in numbers:
    if num > 10:
        large_numbers.append(num)

print(f"All numbers: {numbers}")
print(f"Numbers > 10: {large_numbers}")
````

## 3. The range() Function

Sometimes we want to repeat something a specific number of times, or count through numbers. The `range()` function creates sequences of numbers.

### Three Ways to Use range()

1. `range(stop)` - Counts from 0 up to (but not including) stop
2. `range(start, stop)` - Counts from start up to (but not including) stop
3. `range(start, stop, step)` - Counts from start to stop, incrementing by step

````python
# range(stop) - count from 0
print("Counting 0 to 4:")
for i in range(5):
    print(i)

print()  # Blank line

# This is useful for repeating something N times
print("Printing 'Hello' 3 times:")
for i in range(3):
    print("Hello!")
````

````python
# range(start, stop) - count from start to stop-1
print("Counting 1 to 5:")
for i in range(1, 6):
    print(i)

print()

# Useful for specific number ranges
print("Years 2020 to 2024:")
for year in range(2020, 2025):
    print(year)
````

````python
# range(start, stop, step) - count by different amounts
print("Even numbers 0 to 10:")
for i in range(0, 11, 2):
    print(i)

print()

print("Counting down from 10 to 1:")
for i in range(10, 0, -1):
    print(i)
print("Blastoff!")
````

### Using range() with List Indices

We can use range() with len() to access both the index and the value.

````python
# Access items by index using range
students = ["Alice", "Bob", "Charlie", "Diana"]

print("Class roster:")
for i in range(len(students)):
    print(f"{i + 1}. {students[i]}")

print()

# Modify list items using indices
scores = [85, 92, 78, 95, 88]
print(f"Original scores: {scores}")

# Add 5 bonus points to each score
for i in range(len(scores)):
    scores[i] = scores[i] + 5

print(f"With bonus: {scores}")
````

## 4. Iterating Through Strings

Strings are sequences too! We can loop through each character in a string.

````python
# Loop through each character
word = "Python"

print("Letters in 'Python':")
for letter in word:
    print(letter)

print()

# Count specific characters
text = "Hello World"
vowel_count = 0

for char in text:
    if char.lower() in "aeiou":
        vowel_count = vowel_count + 1

print(f"Text: {text}")
print(f"Number of vowels: {vowel_count}")
````

````python
# Build new strings character by character
original = "python"
uppercase = ""

for char in original:
    uppercase = uppercase + char.upper()

print(f"Original: {original}")
print(f"Uppercase: {uppercase}")

print()

# Remove vowels from a string
text = "Hello World"
no_vowels = ""

for char in text:
    if char.lower() not in "aeiou":
        no_vowels = no_vowels + char

print(f"Original: {text}")
print(f"Without vowels: {no_vowels}")
````

## 5. Nested For Loops

Just like we can nest if statements, we can nest for loops. The inner loop runs completely for each iteration of the outer loop.

### Understanding Nested Loops

````python
# Simple nested loop example
print("Multiplication patterns:")
for i in range(1, 4):  # Outer loop: 1, 2, 3
    for j in range(1, 4):  # Inner loop: 1, 2, 3
        print(f"{i} x {j} = {i * j}")
    print()  # Blank line after each outer loop iteration
````

````python
# Creating a grid pattern
print("Grid of coordinates:")
for row in range(3):
    for col in range(4):
        print(f"({row},{col})", end=" ")
    print()  # New line after each row
````

````python
# Practical example: Comparing items from two lists
colors = ["red", "blue", "green"]
sizes = ["small", "medium", "large"]

print("All possible combinations:")
for color in colors:
    for size in sizes:
        print(f"{size} {color} shirt")
````

## 6. Practical Patterns with For Loops

Let's look at common patterns you'll use frequently.

### Pattern 1: Accumulating a Total

````python
# Calculate sum (even though sum() function exists)
prices = [12.99, 8.50, 15.75, 22.00, 6.25]

total = 0
for price in prices:
    total = total + price

print(f"Prices: {prices}")
print(f"Total: ${total:.2f}")
print(f"Average: ${total / len(prices):.2f}")
````

### Pattern 2: Counting Items That Match Criteria

````python
# Count passing scores
scores = [85, 92, 55, 78, 95, 63, 88, 45, 91]

passing_count = 0
failing_count = 0

for score in scores:
    if score >= 60:
        passing_count = passing_count + 1
    else:
        failing_count = failing_count + 1

print(f"Total tests: {len(scores)}")
print(f"Passing: {passing_count}")
print(f"Failing: {failing_count}")
print(f"Pass rate: {passing_count / len(scores) * 100:.1f}%")
````

### Pattern 3: Finding Maximum or Minimum

````python
# Find the highest score and who got it (parallel lists)
students = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
scores = [85, 92, 78, 95, 88]

# Start by assuming first student has highest score
highest_score = scores[0]
top_student = students[0]

# Check each student
for i in range(len(students)):
    if scores[i] > highest_score:
        highest_score = scores[i]
        top_student = students[i]

print(f"Top student: {top_student}")
print(f"Highest score: {highest_score}")
````

### Pattern 4: Processing Parallel Lists

````python
# Calculate total cost from items and prices
items = ["apple", "banana", "orange", "grape"]
prices = [1.50, 0.75, 2.00, 3.50]
quantities = [3, 5, 2, 1]

print("Shopping Cart:")
total_cost = 0

for i in range(len(items)):
    item_total = prices[i] * quantities[i]
    total_cost = total_cost + item_total
    print(f"{quantities[i]} {items[i]} @ ${prices[i]:.2f} = ${item_total:.2f}")

print(f"\nTotal: ${total_cost:.2f}")
````

## 7. Putting It All Together - Complete Programs

Let's build complete programs that use multiple concepts.

````python
# Grade analyzer
print("=== Grade Analyzer ===")
print()

scores = [85, 92, 78, 95, 88, 73, 91, 67, 82, 90]
print(f"Scores: {scores}")
print()

# Calculate statistics
total = 0
for score in scores:
    total = total + score
average = total / len(scores)

# Count grades by category
a_count = 0
b_count = 0
c_count = 0
d_count = 0
f_count = 0

for score in scores:
    if score >= 90:
        a_count = a_count + 1
    elif score >= 80:
        b_count = b_count + 1
    elif score >= 70:
        c_count = c_count + 1
    elif score >= 60:
        d_count = d_count + 1
    else:
        f_count = f_count + 1

# Display results
print("Statistics:")
print(f"Average: {average:.1f}")
print(f"Highest: {max(scores)}")
print(f"Lowest: {min(scores)}")
print()

print("Grade Distribution:")
print(f"A's: {a_count}")
print(f"B's: {b_count}")
print(f"C's: {c_count}")
print(f"D's: {d_count}")
print(f"F's: {f_count}")
````

````python
# Pattern printer
print("=== Pattern Printer ===")
print()

# Pattern 1: Right triangle
print("Right Triangle:")
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()  # New line

print()

# Pattern 2: Multiplication table
print("5x5 Multiplication Table:")
for i in range(1, 6):
    for j in range(1, 6):
        product = i * j
        print(f"{product:3}", end=" ")  # :3 means 3 characters wide
    print()  # New line after each row
````

## 8. Key Takeaways

**For Loop Basics:**
- `for item in list:` processes each item in a list
- Always use a colon `:` and indent the loop body
- The loop variable (like `item`) is automatically updated each iteration

**Using range():**
- `range(5)` gives 0, 1, 2, 3, 4
- `range(1, 6)` gives 1, 2, 3, 4, 5
- `range(0, 10, 2)` gives 0, 2, 4, 6, 8
- Use `range(len(list))` to get indices

**Common Patterns:**
- Accumulating totals: start with 0, add in the loop
- Counting matches: start with 0, add 1 when condition is true
- Building new lists: start empty, append in the loop
- Processing strings: loop through each character

**Nested Loops:**
- Inner loop completes fully for each outer loop iteration
- Useful for grids, combinations, and comparing lists
- Be careful with indentation!
