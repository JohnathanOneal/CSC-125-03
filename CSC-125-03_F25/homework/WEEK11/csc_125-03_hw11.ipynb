## Part 5: Finding Values in Dictionaries

Practice searching for specific values.

```python
# Example: Find highest value
scores = {
    "Alice": 92,
    "Bob": 85,
    "Charlie": 95,
    "Diana": 88
}

highest_score = 0
top_student = ""

for name, score in scores.items():
    if score > highest_score:
        highest_score = score
        top_student = name

print(f"Top student: {top_student}")
print(f"Highest score: {highest_score}")
```

## Part 6: Building Dictionaries

Create dictionaries by processing lists or counting items.

```python
# Example: Count word occurrences
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]

word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(f"Word counts: {word_count}")
```

## Part 7: Nested Dictionaries

Work with dictionaries that contain other dictionaries.

```python
# Example: Nested student records
students = {
    "Alice": {
        "age": 20,
        "major": "CS",
        "gpa": 3.8
    },
    "Bob": {
        "age": 19,
        "major": "Math",
        "gpa": 3.6
    }
}

print("Student Records:")
for name, info in students.items():
    print(f"{name}: {info['major']}, GPA: {info['gpa']}")
```
