## 1. Why While Loops?

Unlike `for` loops, which run a fixed number of times, **while loops** repeat until a condition becomes false.

Example: run a process until a variable reaches a certain value.

```python
# Basic while example
count = 1
while count <= 5:
    print(f"Count: {count}")
    count += 1
print("Done!")
```

## 2. Input Validation

Use a `while` loop when you don't know how many times you'll need to repeat an action, such as asking for valid input.

```python
# Ask user for a valid number between 1 and 10
num = int(input("Enter a number (1 to 10): "))
while num < 1 or num > 10:
    print("Invalid input. Try again.")
    num = int(input("Enter a number (1 to 10): "))
print(f"You entered: {num}")
```

## 3. Break and Continue

`break` exits the loop early.  
`continue` skips to the next iteration.

```python
# Demonstrating break and continue
n = 0
while n < 10:
    n += 1
    if n == 5:
        continue  # skip printing 5
    if n == 8:
        break     # stop when reaching 8
    print(n)
print("Loop ended.")
```

## 4. Common Error Types

When debugging, always read error messages carefully.

| Error Type | Cause | Example |
|-------------|-------|----------|
| `SyntaxError` | Invalid Python syntax | Missing `:` or parentheses |
| `TypeError` | Wrong type for operation | `'a' + 5` |
| `ValueError` | Invalid value | `int('cat')` |
| `ZeroDivisionError` | Dividing by zero | `5 / 0` |
| `LogicError` | Code runs but gives wrong result | wrong formula or misplaced variable |

## 5. Try / Except - Handling Errors Gracefully

Use `try` / `except` to prevent your program from crashing when something unexpected happens.

```python
# Handling specific errors
try:
    x = int(input("Enter a number: "))
    print(10 / x)
except ValueError:
    print("Please enter a valid integer.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
```

### Catching All Exceptions (for debugging only)
You can catch any error type using a general `Exception` block. This is useful for debugging, but you should normally handle specific errors.

```python
# Catching all exceptions
try:
    user_input = input("Enter something: ")
    result = 10 / int(user_input)
    print(result)
except Exception as e:
    print(f"Something went wrong: {e}")
```

## 6. Debugging in Practice

Example: a small program with an error, then fixed with try/except.

```python
# Problem: this code crashes on 'ten' or 0
nums = ["5", "ten", "15", "0"]
total = 0
count = 0

for n in nums:
    try:
        total += int(n)
        count += 1
    except Exception as e:
        print(f"Skipping '{n}' due to error: {e}")

print(f"Average of valid numbers: {total / count:.1f}")
```

---

## 7. Key Takeaways
- Use `while` when repetition count is not fixed.
- Always change your loop variable to avoid infinite loops.
- `break` exits early; `continue` skips to the next loop.
- Read error messages carefully to find the problem line.
- Use `try/except` to handle predictable errors.
- Use `Exception as e` for broad debugging, but prefer specific exceptions in real code.
