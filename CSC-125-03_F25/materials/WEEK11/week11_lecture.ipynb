## 1. The Problem with Parallel Lists

Remember storing related data in multiple lists?

```python
# Parallel lists - hard to keep in sync
names = ["Alice", "Bob", "Charlie"]
ages = [20, 19, 21]
majors = ["CS", "Math", "CS"]

# To get Bob's info, need to find his index
index = names.index("Bob")
print(f"{names[index]}, {ages[index]}, {majors[index]}")

# What if lists get out of sync?
# This is messy and error-prone!
```

## 2. Introduction to Dictionaries

**Dictionaries** store key-value pairs. Think of a real dictionary: you look up a word (key) to get its definition (value).

Syntax: `{key: value, key: value, ...}`

```python
# Much better - all info together!
student = {
    "name": "Alice",
    "age": 20,
    "major": "CS"
}

print(student)
print()

# Access using keys
print(f"Name: {student['name']}")
print(f"Age: {student['age']}")
print(f"Major: {student['major']}")
```

## 3. Creating Dictionaries

```python
# Method 1: All at once
car = {
    "make": "Toyota",
    "model": "Camry",
    "year": 2020
}
print(f"Car: {car}")
print()

# Method 2: Start empty, add items
book = {}
book["title"] = "Python Basics"
book["author"] = "John Doe"
book["pages"] = 300
print(f"Book: {book}")
```

## 4. Modifying Dictionaries

Dictionaries are mutable - you can change them!

```python
student = {
    "name": "Bob",
    "age": 19,
    "major": "Math"
}
print(f"Original: {student}")

# Change existing value
student["age"] = 20
print(f"After birthday: {student}")

# Add new key-value pair
student["gpa"] = 3.8
print(f"Added GPA: {student}")
```

## 5. Looping Through Dictionaries

```python
grades = {
    "Alice": 92,
    "Bob": 85,
    "Charlie": 78,
    "Diana": 95
}

# Loop through keys
print("Method 1: Keys only")
for name in grades:
    print(f"{name}: {grades[name]}")
print()

# Loop through values only
print("Method 2: Values only")
for grade in grades.values():
    print(f"Grade: {grade}")
print()

# Loop through key-value pairs (MOST COMMON)
print("Method 3: Key-value pairs")
for name, grade in grades.items():
    print(f"{name} earned {grade}")
```

```python
# Calculate class average
grades = {"Alice": 92, "Bob": 85, "Charlie": 78, "Diana": 95}

total = 0
for grade in grades.values():
    total += grade

average = total / len(grades)
print(f"Class average: {average:.1f}")
```

## 6. Essential Dictionary Methods

```python
student = {"name": "Eve", "age": 22, "city": "Boston"}

print(f"Dictionary: {student}")
print()

# Check if key exists
if "name" in student:
    print("'name' exists")

if "email" in student:
    print("'email' exists")
else:
    print("'email' does NOT exist")
print()

# .get() - safe way to access
email = student.get("email", "No email")
print(f"Email: {email}")
print()

# .keys(), .values(), .items()
print(f"Keys: {list(student.keys())}")
print(f"Values: {list(student.values())}")
print(f"Items: {list(student.items())}")
```

## 7. Practical Example - Word Counter

```python
# Count how many times each word appears
text = "hello world hello python world hello"
words = text.split()

word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(f"Text: {text}")
print(f"Word counts: {word_count}")
print()

for word, count in word_count.items():
    print(f"'{word}' appears {count} time(s)")
```

## Key Takeaways

**Creating:**
- `my_dict = {key: value, key: value}`
- `my_dict[key] = value` to add/update

**Accessing:**
- `my_dict[key]` to get value
- `my_dict.get(key, default)` for safe access

**Looping:**
- `for key in my_dict:` - loop through keys
- `for value in my_dict.values():` - loop through values
- `for key, value in my_dict.items():` - loop through pairs

**Checking:**
- `if key in my_dict:` - check if key exists
- `len(my_dict)` - number of key-value pairs
