## Part 5: String Iteration

Loop through strings character by character.

```python
# Example: Count vowels
word = "programming"
vowel_count = 0

for char in word:
    if char.lower() in "aeiou":
        vowel_count = vowel_count + 1

print(f"The word '{word}' has {vowel_count} vowels")
```

## Part 6: Accumulating and Counting

Use loops to accumulate totals and count items.

```python
# Example: Count and sum
numbers = [10, 25, 30, 15, 20]

total = 0
count = 0

for num in numbers:
    total = total + num
    count = count + 1

average = total / count
print(f"Total: {total}")
print(f"Count: {count}")
print(f"Average: {average}")
```
